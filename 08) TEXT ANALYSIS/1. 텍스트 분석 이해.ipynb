# NLP이냐 텍스트 분석이냐?
#### 머신러닝이 보편화되면서 NLP와 텍스트 분석을 구분하는 것이 큰 의미는 없지만, 굳이 구분하자면 NLP는 머신이 인간의 언어를 이해하고 해석하는데 더 중점을 두고, 텍스트 마이닝이라고 불리는 텍스트 분석은 비정형 텍스트에서 의미 있는 정보를 추출하는데 더 중점을 둔다.
#### 예를 들어, NLP는 기계 번역, 자동 질의응답 시스템 같은 것들이 있고, NLP는 텍스트 분석을 향상하게 하는 기반 기술이다. 
#### 기존의 룰 기반 시스템에서 머신러닝을 기반하여 모델을 학습하고 예측하는 시스템으로 기술이 발전하여 NLP와 텍스트 분석의 발전 근간에는 머신러닝이 존재한다고 할 수 있다.
#### 텍스트 분석은 머신러닝, 언어 이해, 통계 등을 활용해 모델을 수립하고 정보를 추출하는 분석 작업을 주로 수행한다.
#### 머신러닝 기술에 힘입어 텍스트 분석은 크게 발전하고 있으며 아래와 같은 영역으로 집중해서 발전해오고 있다.


# 텍스트 분석 이해
#### 텍스트 분석은 비정형 데이터인 텍스트를 분석하는 것이다. 지금까지는 ML 모델에 주어진 정형 데이터기반에서 모델을 수립하고 예측을 수행했다.
#### 하지만, 머신러닝 알고리즘은 숫자형의 피처 기반의 데이터만 입력받을 수 있기 때문에 비정형 텍스트 데이터를 어떻게 피처 형태로 추출하고 피처에 의미 있는 값을 부여하는 것이 핵심 요소이다.
#### 텍스트를 word 기반의 다수의 피처로 추출하고 이 피처에 단어 빈도수와 같은 숫자 값을 부여하면 텍스트는 단어의 조합인 벡터값으로 표현될 수 있는데, 이렇게 변환하는 것이 피처 벡터화 또는 피처 추출이다.
#### 대표적으로 텍스트를 피처 벡터화하는 방법에는 BOW(Bag of Words)와 Word2Vec 방법이 있고, BOW에 대해서 뒤에서 설명할 것이다.
