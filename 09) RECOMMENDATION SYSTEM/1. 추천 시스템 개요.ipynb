# 추천 시스템의 개요
#### 추천 시스템은 결국 사용자 자신도 좋아하는지 몰랐던 취향을 시스템이 발견하고 그에 맞는 콘텐츠를 추천해주는 것이다.
#### 전자 상거래 업체, 영화/음악 콘텐츠 제공 업체 등 많은 곳에서 추천 시스템을 사용하고 있다.

## 온라인 스토어의 필수 요소, 추천 시스템
#### 추천 시스템은 사용자가 무엇을 원하는지 빠르게 찾아내 사용자의 온라인 쇼핑의 즐거움을 배가 되게 합니다.


#### 온라인 스토어는 많은 양의 고객과 상품 관련 데이터를 가지고 있다. 그래서, 이 모든 데이터가 사용자가 흥미를 가질 만한 상품을 즉각적으로 추천하는데 사용한다.


## 추천 시스템의 유형
#### 추천 시스템은 크게 콘텐츠 기반 필터링 방식과 협업 필터링 방식을 나뉜다. 그리고, 협업 필터링 방식은 최근접 이웃 협업 필터링과 잠재 요인 협업 필터링으로 나뉜다.
#### 추천 시스템의 초장기에는 콘텐츠 기반 필터링이나 최근접 이웃 기반 협업 필터링을 주로 사용됐지만, 최근에는 행렬 분해 기법을 이용한 잠재 요인 협업 필터링을 사용한다.
#### 하지만, 서비스하는 아이템의 특성에 따라 다른 필터링 방식도 사용한다.
#### 요즘에는 개인화 특성을 좀 더 강화하기 위해 하이브리드 형식으로 콘텐츠 + 협업 기반을 적절히 결합해 사용하는 경우가 늘고 있다.
